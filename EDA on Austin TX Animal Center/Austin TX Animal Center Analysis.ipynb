```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.mode.chained_assignment = None
```

```python
df_in = pd.read_csv('Austin_Animal_Center_Intakes.csv', parse_dates=['DateTime','MonthYear'])

df_in
```

```python
df_in.shape
```

```python
df_out = pd.read_csv('Austin_Animal_Center_Outcomes.csv', parse_dates=['DateTime','MonthYear'])

df_out
```

```python
df_out.shape
```

# Cleaning Data

#### Duplicate Rows Deleted

```python
df_in = df_in.drop_duplicates()
```

```python
df_in.shape
```

```python
df_out = df_out.drop_duplicates()
```

```python
df_out.shape
```

#### DateTime column deleted from both dfs since it is same as MonthYear column

```python
df_in.drop('DateTime', inplace= True, axis=1)
df_out.drop('DateTime', inplace= True, axis=1)
```

```python
df_in.shape
```

```python
df_out.shape
```

#### Columns with only 1 unique value is deleted

```python
def unique_val(df):
    for this_col in df.columns:
        if(df[this_col].nunique() == 1):
            df = df.drop([this_col], axis= 1)
    return df
```

```python
unique_val(df_in)
df_in.shape
```

```python
unique_val(df_out)
df_out.shape
```

# Dataframes Merged

```python
df = pd.merge(df_in, df_out, on=['Animal ID', 'Name', 'Animal Type', 'Breed', 'Color'], suffixes=('_In', '_Out'))
df
```

```python
df.info()
```

```python
def age_in_months(age):
    if age == 'nan':
        return age
    
    value = age.split(' ')
   
    if value[1] == 'week' or value[1] == 'weeks':
        months = 0
        
    if value[1] == 'month' or value[1] == 'months':
        months = 1
    
    if value[1] == 'day' or value[1] == 'days':
        months = 0
    
    else: 
        months = float(value[0])

    
    return months
```

```python
df['Age upon Intake'] = df['Age upon Intake'].apply(age_in_months)
```

```python
df['Age upon Outcome'] = df['Age upon Outcome'].astype(str)
df['Age upon Outcome'] = df['Age upon Outcome'].apply(age_in_months)
```

```python
df
```

#### Heatmap (Visualization) of Null values

```python
plt.subplots(figsize= (15,5))
sns.heatmap(df.isnull(), cbar= False, cmap= "viridis")
plt.show()
```

# Handling Missing Values

#### Checking how many values are missing in each column

```python
df.isnull().sum().sort_values()
```

#### Checking Percentage of Missing Values

```python
for this_col in df.columns:
    prct = np.mean(df[this_col].isnull())
    print('{} - {}%'.format(this_col, round(prct*100)))
```

### Filling Missing Values

#### Replacing the missing values in a column with "No Name"

```python
df['Name'] = df['Name'].fillna(value= 'No Name')
```

#### Filling the missing value in a column with "Unknown"

```python
df['Outcome Subtype'] = df['Outcome Subtype'].fillna(value= 'Unknown')
```

#### Replacing the missing value with most frequent value using mode

```python
df['Outcome Type'] = df['Outcome Type'].fillna(df['Outcome Type'].mode().iloc[0])
```

#### Calculating Age upon outcome from DOB and MonthYear_Out

```python
for element in (df.index):
    value= df['Age upon Outcome'][element]
    if pd.isnull(value):
        df['Age upon Outcome'][element]= df['MonthYear_Out'][element].year - df['Date of Birth'][element].year - (df['MonthYear_Out'][element].month < df['Date of Birth'][element].month)
```

#### Filling the missing value with nearest value in that column using ffill method

```python
df['Sex upon Intake']= df['Sex upon Intake'].fillna(method='ffill')
df['Sex upon Outcome']= df['Sex upon Outcome'].fillna(method='ffill')
```

#### Heatmap (Visualization) for Null values
##### No Null Values found now

```python
plt.subplots(figsize= (15,5))
sns.heatmap(df.isnull(), cbar= False, cmap= "viridis")
plt.show()
```

# Sanity Check

#### Sanity Check for Breed and Color

```python
import string

alpha= string.ascii_letters
spchar= ",/ -."

allowed_char = alpha+ spchar

for element in df.index:
    for ch in df['Breed'][element]:
        if ch not in allowed_char:
            print('Breed name is not correct')
            print('Sanity check failed')
            break
print("Sanity check for Breed Names done.")
```

```python
alpha= string.ascii_letters
spchar= ",/ -"

allowed_char = alpha+ spchar

for element in df.index:
    for ch in df['Color'][element]:
        if ch not in allowed_char:
            print('Color name is not correct')
            print('Sanity check failed')
            break
print("Sanity check for Color Names done.")
```
